```python
# Importing dataset
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Loading the dataset
data = load_breast_cancer() 

# Converting to DataFrame, DFs offer efficient, flexible data manipulation and integrate well with various libraries
df = pd.DataFrame(data.data, columns=data.feature_names)

# Overview of the dataset
print(df.columns)
print(df.shape)
print(df.head())
print(df.info())
```

```python
print(data.DESCR) # Description of the dataset
"""
- class:
            - WDBC-Malignant 
            - WDBC-Benign
"""
```

```python
# Addding TARGET to DataFrame as 0 -malignant- and 1 -benign-, 'target' column in data already contains the values 0 and 1
df['target'] = data.target

print(df)
print(df.describe())
print(df.info())
```

```python
# 1. Linear Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Defining independent variables, all features except 'mean radius' are independent variables
X = df.drop(['mean radius'], axis=1)  # axis=1 refers to columns and axis=0 refers to rows; .drop function used for removing mean_radius column -dependent variable- on X

# Defining the dependent variable
y = df['mean radius']  # We are trying to predict 'mean radius' based on other variables, so created y with 'mean radius' column

# Splitting the data into training and test sets
# `X_train1` and `X_test1` are the training and testing datasets for the independent variables (features), while `y_train1` and `y_test1` are the corresponding training and testing datasets for the dependent variable (target/'mean radius')
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size=0.2, random_state=1) # 20% of the data will be used for testing, 80% for training

# Creating a Linear Regression model
model1 = LinearRegression()

# Fitting the model to the training data
model1.fit(X_train1, y_train1)

# Now the model can be used to predict 'radius_mean' for the test set
y_pred1 = model1.predict(X_test1)
```

```python
# Evaluating the model 
import statsmodels.api as sm # "api" here refers to a set of classes, methods, and functions provided by a library
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

print("Mean Absolute Error (MAE): ", mean_absolute_error(y_test1, y_pred1))
print("Mean Squared Error (MSE): ", mean_squared_error(y_test1, y_pred1))
print("R-squared: ", r2_score(y_test1, y_pred1))
print("Mean Absolute Percentage Error (MAPE): ", np.mean(np.abs((y_test1 - y_pred1) / y_test1)) * 100)
print("Mean Bias Deviation (MBD): ", np.mean((y_test1 - y_pred1) / y_test1) * 100)
print("Root Mean Square Error (RMSE): ", np.sqrt(mean_squared_error(y_test1, y_pred1)))
print("\nResiduals: ", y_test1 - y_pred1)

# Coefficients
coeff_df1 = pd.DataFrame(model1.coef_, X.columns, columns=['Coefficient'])
print(coeff_df1)

# Fitting the model using statsmodels to see OLS Reg results
X2 = sm.add_constant(X_train1)
est = sm.OLS(y_train1, X2)
est2 = est.fit()
print(est2.summary())

"""
The Ordinary Least Squares (OLS) method estimates the parameters of a linear regression model by minimizing the sum of the squared differences 
between the observed values and the predicted values and is the default method used in the `statsmodels` library and most other statistical packages.
"""
```

**MAE = (1/n) * Σ|y_train - y_pred|;** is the average distance between predicted and actual values.

**MSE = (1/n) * Σ(y_true - y_pred)^2;** is the average of the squared differences between predicted and actual values. Gives more weight to larger errors.

The low MAE and MSE values of this model indicate that the performance of this model is good.

Also, the model explains approximately 99.96% of the variance in the target variable according to R-squared value. Very close to 1.

**MAPE (Mean Absolute Percentage Error)** measures the average percentage difference between the predicted values and the actual values. A lower MAPE indicates better performance. MAPE value of this model is approximately 31.51%, which suggests that the model's predictions deviate from the actual values by around 31.51% on average. This value is relatively high.

**MBD (Mean Bias Deviation)** measures the average bias in the predictions. A positive value indicates the model is over-predicting on average, while a negative value indicates under-predicting. A lower MBD indicates less bias in the model. MBD = 0.0862 (or 8.62%) shows a slight positive bias, meaning the model tends to over-predict the target values by 8.62% on average. This is not very large but indicates a consistent bias that could be adjusted.

**RMSE (Root Mean Squared Error)** measures the square root of the average squared differences between the predicted values and the actual values. It provides an indication of the model's overall error magnitude. A lower RMSE indicates better performance. RMSE value of this model is approximately 0.063, which is relatively low, suggesting that the model's predictions are generally close to the actual values.

**Residuals,** i.e. the differences between actual values and predicted values, are low. But there may be some data points where the model does not perform well.

**The coefficients** expressing how a feature affects the target show that some features have a positive effect and some have a negative effect. For example, an 1 unit increase in the concave point error feature leads to a 3.808430 units increase in radius mean. All other features are assumed to be held constant. 

**ORS regression results** can be used to obtain p-values of t-statistic -coefficients divided by the standard deviation- and many other statistical information.

```python
import statsmodels.api as sm
import numpy as np

summary_df = pd.DataFrame(est2.summary().tables[1].data)

# Setting the column names to the first row of the DataFrame
summary_df.columns = summary_df.iloc[0] # .iloc is used for integer-location based indexing for selection by position
summary_df = summary_df.drop(0) # to avoid duplicate row 

# Converting columns to numeric 
summary_df['coef'] = pd.to_numeric(summary_df['coef'])
summary_df['P>|t|'] = pd.to_numeric(summary_df['P>|t|'])

# Filtering the DataFrame for rows with p-value < 0.05
significant_df = summary_df[summary_df['P>|t|'] < 0.05]

# Sorting the DataFrame by the 'coef' column in descending order
sorted_df = significant_df.sort_values(by='coef', ascending=False)

# Renaming the 95% CI colums to look better
sorted_df = sorted_df.rename(columns={'[0.025': 'lower_bound', '0.975]': 'upper_bound'})

print(sorted_df)
```

Here is the statistically significant features are filtered and organised in descending order of coefficients. 

```python
# 2. Linear Regression with less training data
# How does using 40% of the set in training instead of 80% affect the predictions?

X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y, test_size=0.6, random_state=2) # 60% of the data for testing, 40% for training

model2 = LinearRegression()

model2.fit(X_train2, y_train2)

y_pred2 = model2.predict(X_test2)

print("Mean Absolute Error (MAE): ", mean_absolute_error(y_test2, y_pred2))
print("Mean Squared Error (MSE): ", mean_squared_error(y_test2, y_pred2))
print("R-squared: ", r2_score(y_test2, y_pred2))
print("Mean Absolute Percentage Error (MAPE): ", np.mean(np.abs((y_test2 - y_pred2) / y_test2)) * 100)
print("Mean Bias Deviation (MBD): ", np.mean((y_test2 - y_pred2) / y_test2) * 100)
print("Root Mean Square Error (RMSE): ", np.sqrt(mean_squared_error(y_test2, y_pred2)))
print("\nResiduals: ", y_test2 - y_pred2)

coeff_df2 = pd.DataFrame(model2.coef_, X.columns, columns=['Coefficient'])
print(coeff_df2)

# OLS Regression Results
X2 = sm.add_constant(X_train2)
est = sm.OLS(y_train2, X2)
est2 = est.fit()
print(est2.summary())
```

Model 2 was trained with **less data.** As a result, several differences with Model 1 were observed.

Model 1 has lower MAE, MSE, MAPE, and RMSE, indicating it has overall better accuracy and smaller errors.
Model 2 has a higher R-squared and a significantly lower MBD, indicating it explains variance slightly better and has less bias.

The concave points error feature, which was estimated as 3.80 in the first model, was estimated as 5.66 in this model.
The mean compactness feature, which was estimated as -3.67 in the first model, was estimated as -3.62. 

The models show close results in some values, while they show differences in some values. More tests are required to evaluate the performance of the models. 

When the p-values of the T-tests of the coefficients in the OLS Regression Results are analysed, it is seen that the p-values of Model 2 are higher than Model 1 on average. 

These results shows that **having more training data positively affects the significance of the models.**

```python
# Learning Curve
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

train_sizes = np.linspace(0.1, 1.0, 10) * len(X_train1)
train_sizes = train_sizes.astype(int)


def calculate_learning_curves(model1, model2, X_train1, y_train1, X_test1, y_test1, X_train2, y_train2, X_test2, y_test2, train_sizes):
    train_errors1, val_errors1 = [], []
    train_errors2, val_errors2 = [], []
    
    for m in train_sizes:
        # Model 1
        model1.fit(X_train1[:m], y_train1[:m]) # m represents the current training set size being used in each iteration of the loop.
        y_train_predict1 = model1.predict(X_train1[:m])
        y_val_predict1 = model1.predict(X_test1)
        train_errors1.append(mean_squared_error(y_train1[:m], y_train_predict1))
        val_errors1.append(mean_squared_error(y_test1, y_val_predict1))
        
        # Model 2
        model2.fit(X_train2[:m], y_train2[:m])
        y_train_predict2 = model2.predict(X_train2[:m])
        y_val_predict2 = model2.predict(X_test2)
        train_errors2.append(mean_squared_error(y_train2[:m], y_train_predict2))
        val_errors2.append(mean_squared_error(y_test2, y_val_predict2))
    
    return (train_errors1, val_errors1), (train_errors2, val_errors2)

 
# Calculate learning curve data for model1
(train_errors1, val_errors1), (train_errors2, val_errors2) = calculate_learning_curves(model1, model2, X_train1, y_train1, X_test1, y_test1, X_train2, y_train2, X_test2, y_test2, train_sizes)

# Calculate learning curve data for model2
# train_errors2, val_errors2 = calculate_learning_curves(model2, X_train2, y_train2, X_test2, y_test2, train_sizes)

# Plot learning curves
plt.figure(figsize=(14, 7))

# Plot for model1
plt.subplot(1, 2, 1)
plt.plot(train_sizes, train_errors1, 'g-+', linewidth=2, label="Training error")
plt.plot(train_sizes, val_errors1, 'k--*', linewidth=2, label="Validation error")
plt.title('Learning Curve for Model 1', fontsize=14)
plt.xlabel('Training set size', fontsize=12)
plt.ylabel('Mean Squared Error', fontsize=12)
plt.grid(True)
plt.legend(fontsize=14)
plt.xlim([train_sizes.min(), train_sizes.max()]) # upper limit of x-axis is set to the maximum value in the train_sizes array.
plt.ylim([0, max(max(train_errors1), max(val_errors1), max(train_errors2), max(val_errors2))]) # upper limit of y-axis is set to the maximum value among all the maximum values

# Plot for model2
plt.subplot(1, 2, 2)
plt.plot(train_sizes, train_errors2, 'y--^', linewidth=2, label="Training error")
plt.plot(train_sizes, val_errors2, 'c-.d', linewidth=2, label="Validation error")
plt.title('Learning Curve for Model 2', fontsize=14)
plt.xlabel('Training set size', fontsize=12)
plt.ylabel('Mean Squared Error', fontsize=12)
plt.grid(True)
plt.legend(fontsize=14)
plt.xlim([train_sizes.min(), train_sizes.max()])
plt.ylim([0, max(max(train_errors1), max(val_errors1), max(train_errors2), max(val_errors2))]) 

# Add space between plots
plt.subplots_adjust(wspace=0.3)
plt.show()
```

The Mean Squared Error (MSE) is used in the learning curve as a measure of model performance. 

- If the training error is low but the validation error is high, the model is likely **overfitting.**
    (it performs well on the training data but not on unseen data).
- If both errors are high, the model is likely **underfitting.** 
    (it doesn't perform well even on the training data).
- If both errors are low, the model is likely a **good fit.** 
    (it performs well both on the training data and unseen data)


If we evaluate the curves from different angles, 

    Low Training Set Sizes: For both models, the validation error at low training set sizes is initially much higher than the training error. This indicates that the models initially over-fit small training data. However, as the training set size increases, the validation error decreases and converges to the training error.

    Bias: in general, both models have low training error and therefore show low bias. This shows that the models can fit the training data well.

    Convergence: The difference between training and validation errors narrows and stabilises as the training set size increases, leading to convergence between training and validation errors.
    
    Variance: In both models, the training and validation errors are highly convergent and show low variance. This shows that the initial overfitting decreases with increasing training data. This shows that the models generalise well to the new data.
